```python
def csv_github(csv_url:str,csv_filename):
    if os.path.exists(f'data/{csv_filename}'):
        print("CSV file already exists.")
    else:
        os.makedirs('data')
        response = requests.get(csv_url)
        if response.status_code == 200:
            with open(f'data/{csv_filename}', "wb") as file:
                file.write(response.content)
            print("CSV file downloaded successfully.")
        else:
            print("Failed to download CSV file.")
```

```python
csv_url = "https://raw.githubusercontent.com/AshishJangra27/GFG-Hackathon/main/medals.csv"
csv_filename = "medals.csv"
csv_github(csv_url,csv_filename)
```
